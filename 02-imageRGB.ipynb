```python
import cv2

cap = cv2.VideoCapture(0)  # ใช้ 0 สำหรับกล้องในตัว

if not cap.isOpened():
    print("❌ ไม่สามารถเปิดกล้องได้")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        print("❌ อ่านภาพไม่สำเร็จ")
        break

    print("Frame shape:", frame.shape)

    R = frame[:, :, 2]
    G = frame[:, :, 1]
    B = frame[:, :, 0]

    cv2.imshow("Frame", frame)
    cv2.imshow("Red", R)
    cv2.imshow("Green", G)
    cv2.imshow("Blue", B)

    key = cv2.waitKey(1)
    if key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
